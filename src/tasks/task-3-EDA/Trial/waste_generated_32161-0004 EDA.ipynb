### 3.6) OUTLIERS CHECK

#### We will only be checking outliers of Local Units and Quantity of waster produced as data_belong_to is of object type and there is no possibility of outlier in Year

#### 3.6.1) Local Units

```python
data1=data.copy()
```

```python
sns.distplot(data1['Local units'])
```

```python
sns.boxplot(data1['Local units'])
```

```python
data1['Local units'].describe()
```

```python
def outliers_size(Q3,Q1):
    IQR=Q3-Q1
    outliers_lowerbound=Q1-(1.5*IQR)
    outliers_upperbound=Q3+(1.5*IQR)
    print("IQR is",IQR)
    print("Lower bound:",outliers_lowerbound)
    print("Upper bound:",outliers_upperbound)
outliers_size(1397.5,412)
```

```python
a=data1['Local units']
count=0
for i in a:
    if i>1397.5:
        count=count+1
print("total outliers in size are:",count)
percentage_of_sizeoutliers=(count/48)*100 
print("percentage of size outliers are:",percentage_of_sizeoutliers)
```

### STEP 4) EDA

#### 4.1) Relationship between Year and Local Units

```python
data. groupby(['Year']). sum(). plot(kind='pie', y='Local units',autopct='%1.1f%%')
```

#### The local units were almost evenly distributed over the years

#### 4.2) Relationship between Year and Quantity of Waste Produced

```python
data. groupby(['Year']). sum(). plot(kind='pie', y='Quantity of waste produced',autopct='%1.1f%%')
```

#### The waste produced was almost the same over the years

#### 4.3) Relationship between Country and Local Unit

```python
data. groupby(['data_belong_to']). sum(). plot(kind='pie', y='Local units',autopct='%1.1f%%')
```

#### North Rhine-West Falen has the most Local Units

#### 4.4) Relationship between Country and Quantity of Waste Produced

#### 4.5) Coorelation between Numerical Variables

```python
data2=data[["Local units","Quantity of waste produced"]]
```

```python
fig, ax = plt.subplots(figsize=(10,10))  
sns.heatmap(data2.corr(method ='pearson'),annot=True)
```
