# Data Dictionary/Metadata

1. Year: This refers to the calendar year or time period for which the data is being collected and reported. It represents the specific year to which the data corresponds, such as 2020, 2021, etc.
2. Types of Sales Packaging: This likely refers to the different categories or types of packaging used in the industry. Sales packaging includes the packaging materials used for products that are sold to consumers or other businesses. It can encompass various types of containers, such as bottles, boxes, cans, bags, and more, made from different materials like plastic, glass, metal, or paper.
3. Return of Industry Solutions: This term may refer to the amount or quantity of packaging materials that are returned or collected for recycling or proper disposal by industries or businesses. It could represent the industry's efforts to manage and recycle their packaging waste.
4. Return of System Operators: System operators are entities responsible for organizing and managing recycling or waste management systems, often in compliance with regulatory requirements. The "return of system operators" likely refers to the amount of packaging waste collected and managed by these organized systems.
5. Total Returns: This is the sum total of packaging materials returned or collected for recycling or disposal. It combines the returns from both industry solutions and system operators to provide an overall figure for the packaging waste management efforts in a given year.

:## Autocorrelation and Partial Autocorrelation

````python
import statsmodels.api as sm

fig, axes = plt.subplots(1, 2, figsize=(12, 6))


sm.graphics.tsa.plot_acf(agg_df['Return of Industry Solutions'], ax=axes[0])
axes[0].set_title('Return of Industry Solutions ACF')
axes[0].set_xlabel('Lags')
axes[0].set_ylabel('Auto-Correlation')

sm.graphics.tsa.plot_acf(agg_df['Return of System Operators'], ax=axes[1])
axes[1].set_title('Return of System Operators ACF')
axes[1].set_xlabel('Lags')
axes[1].set_ylabel('Auto-Correlation')

plt.suptitle('Autocorrelation of Aggregated data of type of sales packaging by year')

plt.tight_layout()
plt.show()

#For return of Industry Soultions ACF
#it indicates a perfect positive linear relationship between the variable and its past values at a lag of 0
#As the autocorrelation decreases from 1 to -0.5, it suggests tha the variable's current values are becoming less predictable based on its immediate past values.
#When the autocorrelation increases back to 0, it means that there is no significant linear relationship between the variable and its past values at the given lag.
#In other words, the variable's current values are not predictable based on its past values at this lag.

#For Return of System Operators ACF
##it indicates a perfect positive linear relationship between the variable and its past values at a lag of 0
#As the autocorrelation decreases from 1 to -0.5, it suggests tha the variable's current values are becoming less predictable based on its immediate past values.
#we can see at lag of 10, there is a possibility of increasing back to 0.
````

## Composition of the typs of Sale Packaging for each year

````python
grouped_comp = fdf.groupby(['Year', 'Types of Sales Packaging'])['Total Returns'].sum().reset_index()

years = fdf['Year'].unique()
````

````python
grouped_comp.head(10
                  )
````

````python
years
````

````python
for year in years:
    year_data = grouped_comp[grouped_comp['Year'] == year]
    total_returns = year_data['Total Returns'].sum()
    percentages = year_data['Total Returns'] / total_returns * 100
    explode = [0.5 if percentage < 10 else 0 for percentage in percentages]
    plt.figure(figsize=(8, 10))
    plt.pie(year_data['Total Returns'], labels=None, autopct='%1.1f%%',startangle=140,explode=explode, )
    plt.title(f'Composition of Sales Packaging Waste for {year}')
    plt.legend(year_data['Types of Sales Packaging'], title="Sales Packaging", loc="center left", bbox_to_anchor=(1, 0.5))
    plt.axis('equal')
    plt.show()
````

Null Hypothesis (H0): There is no significant difference in "Total Returns" between different types of packaging.

Alternative Hypothesis (Ha): There is a significant difference in "Total Returns" between at least two types of packaging.

````python
td
````

````python
# hypothesis is already explain above.
from scipy.stats import f_oneway
f_statistic, p_value = f_oneway(*[group_data['Total Returns'] for _, group_data in td.groupby('Types of Sales Packaging')])

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in Total Returns between at least two types of packaging.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in Total Returns between different types of packaging.")
````
