# What is the most common 'Waste Type' in the dataset, and what percentage of the total waste does it represent?

```python
# this is used to find the the most common 'Waste Type'
most_common_waste_type = df['Waste Type'].mode()[0]

# this is used to calculate the count of the most common 'Waste Type'
count_most_common = df['Waste Type'].value_counts().max()

# this displays the total count of waste type in the data
total_count = len(df)

# Calculate the percentage
percentage = (count_most_common / total_count) * 100

print("The most common 'Waste Type' is:", most_common_waste_type)
print("It represents", percentage, "% of the total waste.")
```

**Inference:**
Cases from HZVA v. Coatings, adhesives is the most common waste type in this data

# Bivariate Analysis

# Is there a correlation between the 'Year' and 'Waste generated in my own business' columns? Can this relationship be visualized?

```python
# Calculate the correlation between 'Year' and 'Waste generated in my own business'
correlation = df['Year'].corr(df['Waste generated in my own business'])
correlation
```

```python
# Create a scatter plot to visualize the relationship
plt.scatter(df['Year'], df['Waste generated in my own business'], alpha=0.5)
plt.title("Scatter Plot: Year vs. Waste generated in my own business")
plt.xlabel("Year")
plt.ylabel("Waste generated in my own business")

plt.grid(True)

# Show the plot
plt.show()
```

**Inference:**
The correlation between columns is negative (-0.0013768054409583634)

# Is there a relationship between 'Waste generated in my own business' and 'Waste brought in from within the country'? Do businesses that generate more waste also tend to import more waste?

The aim of this is to explore the relationship between 'Waste generated in my own business' and 'Waste brought in from within the country,' and to understand whether there is a connection or correlation between the amount of waste a business generates internally and the amount of waste it imports from within the country.

```python
# used to calculate the correlation betweent the two columns
correlation_2 = df['Waste generated in my own business'].corr(df['Waste brought in from within the country'])
correlation_2
```

```python
# using a scatter plot to visualize the relationship between the two variables
plt.scatter(x = df['Waste generated in my own business'], y = df['Waste brought in from within the country'], alpha=0.5)
# Adding the correlation_2 coefficient as text annotation
plt.text(3, 5, f'Correlation: {correlation_2:.2f}', fontsize=12, color='red')
plt.title('Scatter Plot: Waste generated in my own business vs. Waste brought in from within the country')
plt.xlabel('Waste generated in my own business')
plt.ylabel('Waste brought in from within the country')
plt.grid(True)
plt.show()
```

**Inference:**
There is a positive correlation relationship between the two columns (0.05497) and this suggest that the two variables tend to move in the same direction. 

Therefore, if one variable goes up, the other also goes up and vice-versa.

# Are there any noticeable trends in the 'Output of waste disposal systems' over the years? Can this plotted in a time series graph to visualize if such trend exists?

```python
yearly_output = df.groupby('Year')['Output of waste disposal systems'].mean()

plt.figure(figsize=(10, 6))
yearly_output.plot(kind='line', marker='o')
plt.title('Trend in Output of waste disposal systems over the Years')
plt.xlabel('Year')
plt.ylabel('Average Output')
plt.grid(True)


plt.show()
```

**Inference:** 
- There seems to be a steep decline in the year 2012 and an increase between the year 2014 and 2016

# Do different 'Plant Types' have different error rates ('Errors for evaluation')? Are certain plant types more prone to errors than others?

```python
# Group the data by 'Plant Type' and calculate the average 'Errors for evaluation' for each plant type
average_errors_by_plant = df.groupby('Plant Type')['Errors for evaluation'].mean()
average_errors_by_plant
```

```python

# Create a bar chart to visualize the average error rates by plant type
plt.figure(figsize=(10, 6))
average_errors_by_plant.plot(kind='bar')
plt.title('Average Error Rates by Plant Type')
plt.xlabel('Plant Type')
plt.ylabel('Average Error Rate')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

**Inference:**
- Aalyzing error rates by plant type can help optimize processes or help to improve quality control. Sorting plants ranks higher amongst others and this might suggest that it's more prone to errors
