```python
import json 
import pandas as pd
import numpy as np
```

```python
with open('E:\\mygit\\Berlin-Chapter-Challenge-Waste-Management\\src\\data\\1-raw data\\Geospatial Data\\waste_disposal\\wastedisposalsites_Saxony-Anhalt_DE-ST.json'
          ,'r', encoding='utf-8') as json_file:
    data = json.load(json_file) 
```

```python
new_data_list = []

for entry in data["elements"]:
    new_entry = {
        **entry.get("tags", {}),  
        "type": entry.get("type", None),  
        "id": entry.get("id", None), 
        "lat": entry.get("lat", None),  
        "lon": entry.get("lon", None), 
    }
    new_data_list.append(new_entry)
```

```python
df = pd.DataFrame(new_data_list)
```
